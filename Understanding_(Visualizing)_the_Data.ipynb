```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('breast-cancer.csv')
```

```python
data
```

```python
import seaborn as sns
```

```python
sns.heatmap(data.drop('id',axis=1).corr())
```

## There are more people with Benign (Non Cancerous) outcome

```python
sns.histplot(data['diagnosis'])
```

## Converting target to categorical

```python
data['diagnosis']=data['diagnosis'].map({'M':1,'B':0})
```

```python
data.sample(10)
```

```python
data.drop('id',axis=1,inplace=True)
data.corr()
```

## From above data we can analyze the correlation between features and target

```python
data.corr()['diagnosis'].sort_values(ascending=False)[1:]
```

## Using this plot we can analyze the outliers present in data for every feature

## We can remove outliers as a process of preprocessing to haldle exceptions in the data

```python

for i in data.columns:
  plt.figure(figsize=(3, 3))
  sns.boxplot(x=data[i])
  plt.title(i)
  plt.show()

```

```python

for a in data.columns:
  plt.figure(figsize=(3, 3))
  sns.histplot(data[a])
  plt.title(a)

```

## Skewness in the data

```python
for i in data.columns:
  print(i,data[i].skew())
```
