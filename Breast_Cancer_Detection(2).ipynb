```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/breast-cancer.csv')
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
data['diagnosis'].value_counts()
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data[['diagnosis']])
```

```python
data['diagnosis'].value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.drop('diagnosis',axis=1),data['diagnosis'],test_size=0.3,random_state=42)
```

```python
y_test
```

```python
Input_features=X_train.columns
```

```python
Input_features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
transform=ColumnTransformer([
    ('scale',StandardScaler(),Input_features)
    ],remainder='passthrough')
```

```python
X_train=transform.fit_transform(X_train)
X_test=transform.transform(X_test)
```

```python
X_train=pd.DataFrame(X_train)
X_train
```

```python
X_test=pd.DataFrame(X_test)
X_test
```

### Since the dimensionality(number of features) of Data is too high we can use dimensionality reduction to reduce the dimension

## Applying -PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

## SVM

```python
from sklearn.svm import SVC
model=SVC(kernel='poly')
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score,accuracy_score
y_pred=model.predict(X_test)
print("Accuracy Score : ",accuracy_score(y_test,y_pred))
print("Recall Score = ",recall_score(y_test,y_pred))
print("Precision Score = ",precision_score(y_test,y_pred,average=None))
print("F1 Score = ",f1_score(y_test,y_pred))
```

```python

model=SVC(kernel='rbf')
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Accuracy Score : ",accuracy_score(y_test,y_pred))
print("Recall Score = ",recall_score(y_test,y_pred))
print("Precision Score = ",precision_score(y_test,y_pred,average=None))
print("F1 Score = ",f1_score(y_test,y_pred))
```

```python
model=SVC(kernel='sigmoid')
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Accuracy Score : ",accuracy_score(y_test,y_pred))
print("Recall Score = ",recall_score(y_test,y_pred))
print("Precision Score = ",precision_score(y_test,y_pred,average=None))
print("F1 Score = ",f1_score(y_test,y_pred))
```
